## pysoundtool.filters.filtersignal()

This function allows both Wiener and Band Spectral Subtraction. Here you can explore visuals of the filtering process as well as hear the results as you adjust the parameters.

```python
# adjust path so we can import the package from the parent directory
import os, sys
import inspect
currentdir = os.path.dirname(os.path.abspath(
    inspect.getfile(inspect.currentframe())))
packagedir = os.path.dirname(currentdir)
sys.path.insert(0, packagedir)

# install what is required to use tools:
import pysoundtool as pyst

# what is necessary to play audio files in this notebook:
import IPython.display as ipd
```

```python
noisy_speech_file = '../audiodata/python_traffic.wav'
samps_unfiltered, sr = pyst.loadsound(noisy_speech_file, sr=41000)
ipd.Audio(samps_unfiltered,rate=sr)
```

```python
filter_type = 'wiener'
samps_filtered, sr = pyst.filtersignal(
    audiofile = samps_unfiltered,     
    sr = sr,
    duration_noise_ms=120,
    filter_type = filter_type,
    apply_postfilter = False,
    win_size_ms=20,
    percent_overlap=0.5,
    real_signal = False,
    visualize = True,
    visualize_every_n_windows = 75)
ipd.Audio(samps_filtered,rate=sr)
```

```python
filter_type = 'wiener_postfilter'
samps_filtered, sr = pyst.filtersignal(
    audiofile = samps_unfiltered,     
    sr = sr,
    duration_noise_ms=120,
    filter_type = filter_type,
    apply_postfilter = True,
    win_size_ms=20,
    percent_overlap=0.5,
    real_signal = False,
    visualize = True,
    visualize_every_n_windows = 75)
ipd.Audio(samps_filtered,rate=sr)
```

```python
filter_type = 'band_spectralsubtraction'
samps_filtered, sr = pyst.filtersignal(
    audiofile = samps_unfiltered,     
    sr = sr,
    duration_noise_ms=120,
    filter_type = filter_type,
    win_size_ms=20,
    percent_overlap=0.5,
    phase_radians = True,
    num_bands = 8,
    visualize = True,
    visualize_every_n_windows = 75)
ipd.Audio(samps_filtered,rate=sr)
```
